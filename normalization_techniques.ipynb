###**Normalization**

(1) **Min-Max** normalization performs a linear transformation on the original data in range [0, 1] or [‚àí1, 1].

**applying on these attribtes**: overall, potential, international_reputation,	weak_foot,	skill_moves, pace, shooting, passing, dribbling,	defending,	physic, attacking_crossing, attacking_finishing	,attacking_heading_accuracy,	attacking_short_passing,	attacking_volleys,	skill_dribbling,	skill_curve,	skill_fk_accuracy,	skill_long_passing,	skill_ball_control,	movement_acceleration,	movement_sprint_speed,	movement_agility,	movement_reactions,	movement_balance,	power_shot_power,	power_jumping,	power_stamina,	power_strength,	power_long_shots,	mentality_aggression,	mentality_interceptions,	mentality_positioning,	mentality_vision,	mentality_penalties,	mentality_composure,	defending_marking,defending_standing_tackle,	defending_sliding_tackle,	goalkeeping_diving,	goalkeeping_handling,	goalkeeping_kicking,	goalkeeping_positioning,	goalkeeping_reflexes

--------------------------------------------------------------------------------
All these are various scores of the players which are in different ranges either in 1-100 oe 1-5 so we normalize all of these scores in the range 
0-1

```python
def min_max(attribute):

  min_attribute_val=df[attribute].min()
  max_attribute_val=df[attribute].max()

  new_min=0
  new_max=1

  for i in range(len(df[attribute])):
    df[attribute][i]= (((( df[attribute][i] ) - ( min_attribute_val ))*(new_max-new_min))/(max_attribute_val-min_attribute_val)) + new_min 
    
```

```python
to_be_normalized=['overall', 'potential', 'international_reputation',	'weak_foot','skill_moves', 'pace', 'shooting', 'passing', 'dribbling',	'defending',	'physic', 'attacking_crossing', 'attacking_finishing'	,'attacking_heading_accuracy',	'attacking_short_passing',	'attacking_volleys',	'skill_dribbling',	'skill_curve',	'skill_fk_accuracy','skill_long_passing',	'skill_ball_control',	'movement_acceleration','movement_sprint_speed','movement_agility',	'movement_reactions',	'movement_balance',	'power_shot_power',	'power_jumping','power_stamina','power_strength','power_long_shots','mentality_aggression','mentality_interceptions',	'mentality_positioning',	'mentality_vision',	'mentality_penalties',	'mentality_composure',	'defending_marking','defending_standing_tackle',	'defending_sliding_tackle',	'goalkeeping_diving',	'goalkeeping_handling',	'goalkeeping_kicking',	'goalkeeping_positioning',	'goalkeeping_reflexes']
```

```python
for i in to_be_normalized:
  df=df.astype({i: np.float64})
  min_max(i)
```

(2) **Z-score** normalizes the value for an attribute using the mean and standard
deviation

applying on attributes **age**, **height**, **weight**

```python
def z_score(attribute):

  mean=df[attribute].mean()
  stdev=df[attribute].std()

  for i in range(len(df[attribute])):
    df[attribute][i]= ( df[attribute][i] - mean )/( stdev )   
```

```python
z_score('age')
z_score('height_cm')
z_score('weight_kg')
```

(3) **Decimal scaling**  normalizes the value of attribute by moving the decimal point in
the value.

In this dataset we have **value_eur**, **wage_eur** attributes which are basically total net worth and wage (salaries) of the players and also **release_clause_eur** which is the attribute representing how much money is to be paid in order to buy the club in which the player is.

Since these values are very large in the order of 10^8 and 10^6 so we will normalize these values in the range of [0,1]



applying on attributes **value_eur**, **wage_eur** , **release_clause_eur**

```python
def decimal_scaling(attribute):

  maxx=df[attribute].max()
  j=len(str(maxx))

  for i in range(len(df[attribute])):
    df[attribute][i]=( df[attribute][i] )/( 10**j )
```

```python
decimal_scaling('value_eur')
decimal_scaling('wage_eur')
decimal_scaling('release_clause_eur')
```
